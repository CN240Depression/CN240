```python
import cv2
import os
import numpy as np
import dlib
import pandas as pd
from scipy import spatial
```

```python
def make_csv_real():
    
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor('./shape_predictor_68_face_landmarks.dat')
    option = []
    for file in os.listdir(r'C:\Users\Poott\OneDrive\Desktop\CN240\CN240\images\notblur'): 
        print(file)
        option.append(file)
    for folder in option :  
        for folder_path in os.listdir(r'C:\Users\Poott\OneDrive\Desktop\CN240\CN240\images\notblur'+"/"+folder):
            print(folder_path)
            for file in os.listdir(r'C:\Users\Poott\OneDrive\Desktop\CN240\CN240\images\notblur'+"/"+folder+"/"+folder_path):
                image = cv2.imread(r'C:\Users\Poott\OneDrive\Desktop\CN240\CN240\images\notblur'+"/"+folder+"/"+folder_path+"/"+file)
                image = cv2.resize(image,(512,512))
                # Convert the image color to grayscale
                gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                # Detect the face
                rects = detector(gray, 1)
                
                if not rects :
                    print("test")
                # Detect landmarks for each face
                else:
                    
                    for rect in rects:
                        # Get the landmark points
                        shape = predictor(gray, rect)

                    
                    # Convert it to the NumPy Array
                        shape_np = np.zeros((68, 2), dtype="int")
                        for i in range(0, 68):
                            shape_np[i] = (shape.part(i).x, shape.part(i).y)
                        shape = shape_np
                        
                        # Display the landmarks
                    
                        for i, (x, y) in enumerate(shape):
                            if i == 30:
                                cv2.circle(image, (x, y), 1, (0, 255, 0), 30)
                        # Draw the circle to mark the keypoint
                            elif i <30 and i> 16 :
                                
                                cv2.circle(image, (x, y), 1, (255, 0, 0), 30)

                            else:
                                
                                cv2.circle(image, (x, y), 1, (0, 0, 255), 10)

                    for i, (x, y) in enumerate(shape):
                        if i ==30:
                            x_main = x
                            y_main = y

                    distance = []
                    for i, (x, y) in enumerate(shape):
                        print("x",x)
                        print("y",y)
                        if i >=0 and i <68 and i != 30:
                            result = spatial.distance.cosine((x,y),(x_main,y_main))
                            distance.append(result)

                    index = []
                    for x in range(0,68):
                        if x != 30:
                            index.append(x)

                    dict = {'Path':folder,'index':index,'distance':distance,}
                    path_output = r'C:\Users\Poott\OneDrive\Desktop\CN240\CN240\csv'
                    df = pd.DataFrame(dict)
                    df.to_csv(path_output+'/'+folder+"_csv/"+folder+'.csv', mode='w+', header="test",index=False)                        
            
```

```python
make_csv_real()
```
